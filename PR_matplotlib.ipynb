```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
from yahoo_historical import Fetcher
```

```python
dfINDF = Fetcher('INDF.JK', [2011,1,1],[2019,12,20]).getHistorical()
```

```python
dfINDF.head()
```

```python
dfINDF.isna().any()
```

```python
dfINDF.info()
```

```python
dfINDF = dfINDF.fillna(method = 'ffill')
```

```python
dfINDF.info()
```

```python
dfINDF['Date'] = pd.to_datetime(dfINDF['Date'])
```

```python
dfINDF.info()
```

```python
dfINDF.set_index('Date',inplace = True)
```

```python
dfINDF
```

```python
plt.figure(figsize = (8,8))
plt.plot(dfINDF['Close'])
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize =(20,15))
axes[0][0].plot(dfINDF['Open'],'r', label = 'INDF Open')
axes[0][0].set_xlabel('Year')
axes[0][0].set_ylabel('Rupiah')
axes[0][0].set_title('INDF Open')

axes[0][1].plot(dfINDF['High'],'g',label = 'INDF High')
axes[0][1].set_xlabel('Year')
axes[0][1].set_ylabel('Rupiah')
axes[0][1].set_title('INDF High')

axes[1][0].plot(dfINDF['Volume'],'b',label = 'INDF Volume')
axes[1][0].set_xlabel('Year')
axes[1][0].set_ylabel('Volume in Thousands')
axes[1][0].set_title('INDF Volume')

axes[1][1].plot(dfINDF['Close'],'y', label = 'INDF Close')
axes[1][1].set_xlabel('Year')
axes[1][1].set_ylabel('Rupiah')
axes[1][1].set_title('INDF Volume')

```
